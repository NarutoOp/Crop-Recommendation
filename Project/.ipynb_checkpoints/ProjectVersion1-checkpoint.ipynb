```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
data=pd.read_csv("cropdata.csv",index_col=0)
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
x1=data["State"]
y1=data["Soil"]
z1=data["Month"]
x = le.fit_transform(x1)
y = le.fit_transform(y1)
z = le.fit_transform(z1)
data1 = data.merge(pd.DataFrame({'State_C':x,'Soil_C':y,'Month_C':z}),right_index=True, left_index=True)
data1.head()
```

```python
X = data1[['State_C','Soil_C','Month_C']]
Y = data1[['Rice','Wheat','Cotton','Sugarcane','Tea','Ragi','Maize','Millet','Barley']]
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X,Y)
```

```python
a=classifier.predict([['0','0','4']])
a
```

```python
for i in range(0,9,1):
    if a[0][i] == 1:
        if i == 0:
            print("Rice")
        elif i == 1:
            print("Wheat")
        elif i == 2:
            print("Cotton")
        elif i == 3:
            print("Sugarcane")
        elif i == 4:
            print("Tea")
        elif i == 5:
            print("Ragi")
        elif i == 6:
            print("Maize")
        elif i == 7:
            print("Millet")
        elif i == 8:
            print("Barley")
        
```
